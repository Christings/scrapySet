### 1.2 分词、去停用词和制作区域文本

```python
def split_and_delete_words(path="data/content.txt"):
    df=pd.read_csv(path)
    # df["cut"]=df["content"].apply(lambda x:list(jieba.cut(x)))

    stopwords=pd.read_csv("data/stopwords/stopwords.txt", names=["stopword"], delimiter="\t", quoting=csv.QUOTE_NONE, encoding='utf-8')
    stop_list=stopwords['stopword'].tolist()
    
    df['cut'] = df['content'].apply(lambda x : [i for i in jieba.cut(x) if i not in stop_list])
    
    result=[]
    for index,row in df.iterrows():
        result+=row["cut"]
    
    highwords=Counter(result)
    highwords=dict(highwords)
    
    for k,v in list(highwords.items()):
        if len(k) < 3:
            del highwords[k]
    highwords=Counter(highwords)
    print(highwords.most_common(500))
#     df["cut"].to_csv("data/content_cut.txt", header="content", index=False)
    return highwords
```

## 二、制作词云

```python
def make(df,name):
    text=""
    for index,row in df.iterrows():
        row=row.to_string()
        text+=row
        text+=" "
        
    background=np.array(Image.open('data/balloon.png'))
    fontpath="data/SiYuanSongTiRegular/SourceHanSerifCN-Regular-1.otf"
    
    wc = WordCloud(
    background_color='white',  # 设置背景颜色，与图片的背景色相关
    mask=background,  # 设置背景图片
    font_path=fontpath,  # 显示中文，可以更换字体
    max_words=2000,  # 设置最大显示的字数
    stopwords={"来源","月","元","近日"},  # 设置停用词，停用词则不再词云图中表示
    max_font_size=500,    # 设置字体最大值
    min_font_size=20,     # 字体最小值
    random_state=42,      # 设置有多少种随机生成状态，即有多少种配色方案
    collocations=False,   # 避免重复单词
    scale=1,  # 设置生成的词云图的大小
    width=1600,height=1200,margin=10, #图像宽高，字间距，需要配合下面的plt.figure(dpi=xx)放缩才有效
    )
    
    wc.generate(text)
    
    img_color=ImageColorGenerator(background)
    plt.figure(dpi=100)
    plt.imshow(wc.recolor(color_func=img_color))
    plt.axis("off") # 隐藏坐标
    plt.show()
    
    wc.to_file(name +'.png')
```

```python
make(df_cut,"data/all")
```

```python
make(df_beijing,"data/beijing")
```

```python
make(df_guangdong,"data/guangdong")
```

## 词频直方图

```python
def make_pic(df, name):
    text=[]
    for index,row in df.iterrows():
        text+=row[0].split(" ")
    highwords=Counter(text)
    
    tmp=highwords.most_common(100)
    print(tmp)
    high=highwords.most_common(20)
        
    plt.rcParams['font.sans-serif']=['SimHei']
    params = {
    'figure.figsize': '12, 4'
    }
    plt.rcParams.update(params)
    for i in high:
        
        plt.bar(i[0],i[1])
        plt.text(i[0],i[1]+0.05,"%.0f"%i[1], ha="center", va= 'bottom',fontsize=7)
         

    plt.title("劳动冲突词频top-20")
#     plt.xlabel("单词")
    plt.ylabel("词频")
#     plt.legend("词频直方图")
#     plt.axis('off')  #去除空白区域
    plt.savefig("data/"+name, dpi=1000, bbox_inches="tight", pad_inches=1, transparent = True, facecolor ="w")
#     plt.savefig("squares1.png", 
#             bbox_inches ="tight", 
#             pad_inches = 1, 
#             transparent = True, 
#             facecolor ="g", 
#             edgecolor ='w', 
#             orientation ='landscape') 
  
    plt.show()

make_pic(df_cut,"words_all.png")
```
